## for videos

```python
from ultralytics import YOLO

# Load model
model = YOLO("yolov8n.pt")

# Path to input video
video_path = "Untitled Project.mp4"  # change this

# Run detection on video
model.predict(
    source=video_path,
    save=True,       # save output video
    conf=0.5
)
```
